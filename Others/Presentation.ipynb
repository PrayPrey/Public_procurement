## 위의 경우처럼 outlier을 모두 고려하기에는 문제가 있고, 실제로 저런 특수한 발주가 일어나는 경우까지 시계열로 예측하기에는 어려움이 있어보인다. 패턴인식을 통해서 발주가 일어나는 시기를 예측한다손 치더라도 저런 outlier를 제거하는 것은 반드시 필요한 일일 듯하다.
## 그렇다면 어떤 값을 outlier로 볼 것인가? 발주금액으로만 볼 것인가 아니면,... 일단은 발주금액으로만 보자!

## 발주금액을 예측하는 것은 과연 의미가 있는 것일까? 일단은 이상치를 제외한 발주금액 예상을  보도록 하자.

```python
def t_square(data ,alpha: float = None):
    
    array_data = np.array(data)

    
    p = array_data.shape[1]
    m = array_data.shape[0]
    n = 1
    col_mean= np.array(data.agg(['mean']))
    mat_cov = data.cov()
    mat_inv_s = np.linalg.inv(mat_cov)  # inverse S
    
    values = []
    for sample in array_data:
        dif = sample - col_mean
        value = np.matmul(np.matmul(dif,mat_inv_s),dif.T).ravel().tolist()  # T-square
        values.append(value)
    
    LCL = 0
    UCL = ( ( p * (m+1) * (m-1) ) / (m * (m - p)) ) * f.ppf((1-alpha), p, m-p)
            
    return values, LCL, UCL 
```

```python
import statistics

def ewma(data, k):
    lst = []
    for i in range(len(data)):
        if i == 0:
            alpha = statistics.mean(data)
        else:
            alpha = k*data.iloc[i] +  (1-k)*alpha
        lst.append(round(alpha,2))
    return lst

df_new = pd.DataFrame(ewma(y_test,0.25),columns=['x'])
```

```python
def control_limits(data,k):
    lst = []
    mean = statistics.mean(data['x'])
    for i in range(1,len(data)+1):
        
        w = k/(2-k)
        m = np.sqrt( w * (1- pow(1-k,2*i)))
        ucl = mean + 3 * statistics.stdev(data['x']) * np.sqrt(k/(2-k) * (1-np.power(1-k, 2*i)))
        lcl = mean - 3 * statistics.stdev(data['x']) * np.sqrt(k/(2-k) * (1-np.power(1-k, 2*i)))
        lst.append((ucl,lcl))
    return lst



df_new_ucl = [i[0] for i in control_limits(df_new,0.25)]
df_new_lcl = [i[1] for i in control_limits(df_new,0.25)]


print(control_limits(df_new,0.2))
```
