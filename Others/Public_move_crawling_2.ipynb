```python
import pandas as pd

lst_see = [21, 22, 24, 26, 27, 25,1]
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time


lst_all = []

options = webdriver.ChromeOptions()
options.add_experimental_option('excludeSwitches', ['enable-logging'])
# options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
options.add_argument("lang=ko_KR")
# options.add_argument("User-Agent: 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:65.0) Gecko/20100101 Chrome/65.0")

lst_all = []

for jo in lst_see:
    driver = webdriver.Chrome(options = options)
    driver.get('https://ppstat.g2b.go.kr:8411/pd/report/uesrReportList.do?reportGubun='+str(jo))
    time.sleep(0.1)
    driver.refresh()

    data = driver.find_elements_by_xpath('//*[@id="dd1"]/div') # 해당 목록에 있는 전체 목록의 갯수
    
    for i in range(len(data)):
        element = driver.find_element_by_xpath('//*[@id="dd1"]/div['+ str(i+1) +']/a[2]') # 전체 목록 중 하나
        element.click() # 옆의 정보창 클릭


        driver.switch_to.window(driver.window_handles[1])
        driver.get_window_position(driver.window_handles[1])


        # 정보창 내에서 데이터 불러오기
        time.sleep(0.1)
        small_element_1 = driver.find_element_by_xpath('//*[@id="pop"]/div[2]/div[1]/div[2]/table/tbody/tr[4]/td')
        small_element_2 = driver.find_element_by_xpath('//*[@id="pop"]/div[2]/div[1]/div[2]/table/tbody/tr[1]/td')
        small_element_3 = driver.find_element_by_xpath('//*[@id="pop"]/div[2]/div[1]/div[2]/table/tbody/tr[6]/td')
        small_element_4 = driver.find_element_by_xpath('//*[@id="pop"]/div[2]/div[1]/div[2]/table/tbody/tr[7]/td/p')
        
        lst_all.append([small_element_1.text, small_element_2.text, small_element_3.text, small_element_4.text])


        driver.switch_to.window(driver.window_handles[0])
        driver.get_window_position(driver.window_handles[0])


    driver.quit()
```

```python
df_lst_all = pd.DataFrame(lst_all,columns=['보고서 분류','통계명','보고서요약','통계설명'])
# df_lst_all.to_csv('온통조달.csv',encoding = 'cp949',index = False)
```
