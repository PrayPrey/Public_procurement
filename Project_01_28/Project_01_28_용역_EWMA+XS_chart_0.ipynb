```python
import statistics

def control_limits(data,k):
    lst = []
    mean = statistics.mean(data['x'])
    for i in range(1,len(data)+1):
        
        w = k/(2-k)
        m = np.sqrt( w * (1- pow(1-k,2*i)))
        ucl = mean + 3 * statistics.stdev(data['x']) * np.sqrt(k/(2-k) * (1-np.power(1-k, 2*i)))
        lcl = mean - 3 * statistics.stdev(data['x']) * np.sqrt(k/(2-k) * (1-np.power(1-k, 2*i)))
        lst.append((ucl,lcl))
    return lst



df_new_ucl = [i[0] for i in control_limits(df_배정예산,0.1)]
df_new_lcl = [i[1] for i in control_limits(df_배정예산,0.1)]
```

```python
def make_automation(code , time_interval):
    
    
    df_make_s_chart_2_0 = df_make_s_chart.groupby(['수요기관코드',pd.Grouper(key = '공고게시일자', freq = time_interval)]).agg({'std','count','mean'})
    
    
    
    if sum([i[0] == str(code) for i in df_make_s_chart_2_0.index]) == 0:
        print('없는 기관이거나 최상위 기관의 코드인지를 확인하십시오.')
        return code
 
    
    
    b= df_make_s_chart_2_0[[i[0] == str(code) for i in df_make_s_chart_2_0.index]]['배정예산'].copy()
    
    
    if b['std'].isnull().sum() != 0:
        print('표준편차가 0인 부분이 있습니다.')
        return b

    
    b_1 = b.reset_index()[['mean','std','count']].copy()
    b_2 = b_1.copy()
    
    
    if b_1[['count']].isnull().sum().values[0] != 0:
        print('NaN error')
        return b_1

    test = []
    for i in range(b_1.shape[0]):
        n = b_1.loc[i, 'count']
        std = b_1.loc[i,'std']
        test.append( (n-1)*pow(std,2) )

    w = statistics.sqrt( sum(test) / (b_1.apply(sum)['count'] - b_1.shape[0]) ) 
    
    each_s, s_bar, s_ucl , s_lcl = xs_chart_1(b_1, w)

    df_ucl = pd.DataFrame( s_ucl , columns = ['ucl']).fillna(0)
    df_lcl = pd.DataFrame( s_lcl , columns = ['lcl']).fillna(0)

    fig = go.Figure()

    fig.add_trace(go.Scatter(x = b_1.index ,y = [s_bar] * b_1.shape[0],line_shape='hv',name= 'Mean'))

    fig.add_trace(go.Scatter(x = b_1.index ,y = each_s,mode= 'markers+lines',name= 'Monitoring Statistics',marker_color = 'gray'))

    fig.add_trace(go.Scatter(x = b_1.index ,y = df_lcl['lcl'],mode= 'markers', marker_symbol='line-ew',
                         marker_line_color="blue", marker_color="red",
                           marker_line_width=2, marker_size=15,name= 'LCL'))

    fig.add_trace(go.Scatter(x = b_1.index ,y = df_ucl['ucl'],mode= 'markers',name= 'UCL', marker_symbol='line-ew',
                         marker_line_width=2, marker_size=15,marker_line_color="red", marker_color="blue"))




    fig.update_layout(width=1200, height=600,
                  yaxis_range=[10e+07,300e+07],              
                 xaxis_range=[0,50],
                  title = "This X_S Control (수요기관코드 : " +code+ ") ")
    fig.show()
    
    df_all = pd.concat([df_ucl,pd.DataFrame(each_s),df_lcl],axis=1)
#     return b 
    return df_all

        
```
