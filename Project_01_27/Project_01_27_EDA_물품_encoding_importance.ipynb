```python
import numpy as np # 각 모델에서 내부적으로 관련 라이브러리 사용 가능
import pandas as pd

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

from sklearn.neighbors import KNeighborsRegressor            # 1. K-Nearest Neighbor(KNN)
from sklearn.svm import SVR                                               # 3. SVC
from sklearn.tree import DecisionTreeRegressor                  # 4. Decision Tree
from sklearn.ensemble import RandomForestRegressor       # 5. Random Forest
from sklearn.ensemble import ExtraTreesRegressor            # 6. Extra Tree
from sklearn.ensemble import GradientBoostingRegressor  # 7. GBM

from xgboost import XGBRFRegressor                                    # 9. XGBoost
from lightgbm import LGBMRegressor                              # 10. LightGBM

from tqdm import notebook
import time 
```

```python

knn_model = KNeighborsRegressor(n_jobs = -1)
random_model = RandomForestRegressor(n_jobs=-1)
extra_model = ExtraTreesRegressor(n_jobs=-1)
gbm_model = GradientBoostingRegressor()

xgb_model = XGBRFRegressor(eval_metric='logloss')
lgbm_model = LGBMRegressor()

models = [
    knn_model,
    random_model,
    extra_model,
    gbm_model,
    xgb_model,
    lgbm_model
]
notebook.tqdm
k_fold = KFold(n_splits=5, shuffle=True, random_state=0)           # K-Fold 사용
results = dict()
for alg in notebook.tqdm(models,desc='iterate list'):
    alg.fit(X_train, y_train,)    
#     score = make_scorer(mean_squared_error)
#     score_num = score(alg,X_train,y_train)
    score = cross_val_score(alg, X_train, y_train.values.ravel(), cv=k_fold, scoring='r2',n_jobs=-1)
    results[alg.__class__.__name__] = np.mean(score)*100
```
